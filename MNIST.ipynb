```python
from keras.models import Sequential
from keras.layers import Dense, Activation
from keras.optimizers import SGD
from keras.datasets import mnist
from keras.utils import np_utils
from keras.regularizers import l2
from matplotlib import pyplot as plt

import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
np.seterr(all='raise')
```

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
X_train = X_train.reshape(60000, 784)
X_test = X_test.reshape(10000, 784)

X_train = X_train.astype('float32')
X_test = X_test.astype('float32')

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
m = X_train.shape[0]
k = 10
L = (k - 1) / (k * m) * np.linalg.norm(X_train) #+ 10 * 2.89
print(1/L)
```

```python
Y_train = np_utils.to_categorical(y_train, 10)
Y_test = np_utils.to_categorical(y_test, 10)
```

```python
model = Sequential()
model.add(Dense(10, input_shape=(784,), name='dense', kernel_initializer='zeros', bias_initializer='zeros'))
model.add(Activation('softmax'))
```

```python
sgd = SGD(lr=0.1)

model.compile(loss='categorical_crossentropy', optimizer=sgd, metrics=['accuracy'])
```

```python
history = model.fit(X_train, 
                    Y_train, 
                    epochs=500, 
                    batch_size=128, 
                    validation_data=(X_test, Y_test), 
                    verbose=1)
```

```python
history.history.keys()
```

```python
acc01 = history.history['val_acc']
loss01 = history.history['val_loss']
```

```python
fig, ax = plt.subplots(ncols=2, figsize=(10, 4))

plt.suptitle('Training history with $\\alpha=0.1$')

ax[0].set_ylabel('Validation loss')
ax[0].set_xlabel('Epochs')
ax[0].plot(range(1, 501, 5), loss01[::5], 'r-');

ax[1].set_ylabel('Validation accuracy')
ax[1].set_xlabel('Epochs')
ax[1].plot(range(1, 501, 5), acc01[::5], 'r-');

fig.tight_layout()
fig.subplots_adjust(top=0.88)
```

```python
sgd = SGD(lr=1/L)

model.compile(loss='categorical_crossentropy', optimizer=sgd, metrics=['accuracy'])
```

```python
history2 = model.fit(X_train, 
                    Y_train, 
                    epochs=500, 
                    batch_size=128, 
                    validation_data=(X_test, Y_test), 
                    verbose=1)
```

```python
accL = history2.history['val_acc']
lossL = history2.history['val_loss']
```

```python
fig, ax = plt.subplots(ncols=2, figsize=(10, 4))
plt.suptitle('Training history with $\\alpha=1/L$')

ax[0].set_ylabel('Validation loss')
ax[0].set_xlabel('Epochs')
ax[0].plot(range(1, 501, 10), lossL[::10], 'r-');

ax[1].set_ylabel('Validation accuracy')
ax[1].set_xlabel('Epochs')
ax[1].plot(range(1, 501, 10), accL[::10], 'r-');

fig.tight_layout()
fig.subplots_adjust(top=0.88)
```

```python
fig, ax = plt.subplots(ncols=2, figsize=(10, 4))
plt.suptitle('Comparison of training and validation accuracies with different $\\alpha$')

ax[0].set_ylabel('Training accuracy');
ax[0].set_xlabel('# iterations');
lines = ax[0].plot(range(1, 501, 5), history.history['acc'][::5], 'r-',
                   range(1, 501, 5), history2.history['acc'][::5], 'g-');
ax[0].legend(lines, ('$\\alpha=0.1$', '$\\alpha=1/L$'));

ax[1].set_ylabel('Validation accuracy')
ax[1].set_xlabel('# iterations')
lines = ax[1].plot(range(1, 501, 5), history.history['val_acc'][::5], 'r-',
                   range(1, 501, 5), history2.history['val_acc'][::5], 'g-');
ax[1].legend(lines, ('$\\alpha=0.1$', '$\\alpha=1/L$'))

fig.tight_layout()
fig.subplots_adjust(top=0.88)
```

```python
fig, ax = plt.subplots(ncols=2, figsize=(10, 4), sharey=True)

ax[0].set_ylabel('Accuracy');
ax[0].set_xlabel('# iterations');
lines = ax[0].plot(range(1, 501, 5), history.history['acc'][::5], 'r-',
                   range(1, 501, 5), history.history['val_acc'][::5], 'g-');
ax[0].legend(lines, ('Training', 'Validation'));

ax[1].set_xlabel('# iterations')
lines = ax[1].plot(range(1, 501, 5), history2.history['acc'][::5], 'r-',
                   range(1, 501, 5), history2.history['val_acc'][::5], 'g-');
ax[1].legend(lines, ('Training', 'Validation'))

fig.tight_layout()
```
